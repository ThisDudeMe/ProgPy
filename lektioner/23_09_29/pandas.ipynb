```python
import numpy as np

# Skapa en DataFrame med försäljningsdata och saknade värden
data = {
    'Datum': pd.date_range(start='2024-01-01', periods=5, freq='D'),
    'Sales': [10, np.nan, 15, np.nan, 20],
}
df = pd.DataFrame(data)

# Ta bort rader med saknade värden
df_dropna = df.dropna()

# Fyll saknade värden med 0
df_fillna = df.fillna(0)

# Interpolering av försäljning
df_interpolate = df.copy()
df_interpolate["Sales"] = df["Sales"].interpolate()

# Visa resultaten
print("Original DataFrame:\n", df)
print("\nEfter dropna:\n", df_dropna)
print("\nEfter fillna:\n", df_fillna)
print("\nEfter interpolering:\n", df_interpolate)
```
