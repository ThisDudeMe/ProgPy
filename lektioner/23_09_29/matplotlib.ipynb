```python
import matplotlib.pyplot as plt
```

```python
# Första värdet är x-axeln, andra värdet y-axeln.
plt.plot(['Mon', 'Tues', 'Weds', 'Thurs', 'Fri', 'Sat', 'Sun'], [540, 500, 490, 590, 680, 710, 610])
```

```python
# Det går bra att spara i variabler också.
days = ['Mon', 'Tues', 'Weds', 'Thurs', 'Fri', 'Sat', 'Sun']
hours_worked = [5, 4, 7, 8, 3, 1, 1]
plt.plot(days, hours_worked)
```

```python
# Oftast läser du in data från externa filer. 
# Pandas är ett populärt bibliotek som bland annat används till detta.
import pandas as pd

plants_simple = pd.read_csv("plants_simple.csv")

plants_simple.head(15) #Head visar de första n värdena.
```

```python
# Du kan nu plotta genom att välja olika kolumner.
plt.plot(plants_simple["Time"], plants_simple["Plant_height"])
# Skriver du såhär kommer ytterligare en linje.
plt.plot(plants_simple["Time"], plants_simple["Lateral_spread"])
```

```python
# Du kan också styla dina grafer med color, linewidth och linestyle. 
plt.plot(plants_simple["Time"], plants_simple["Plant_height"], color="g", linewidth=2, linestyle="dotted")
plt.plot(plants_simple["Time"], plants_simple["Lateral_spread"])
```

```python
# Du kan lägga till en beskrivning för varje linje (legend). Du gör det med label-attributet.
plt.plot(plants_simple["Time"], plants_simple["Plant_height"], color="g", linewidth=2, linestyle="dotted", label="Plant height")
plt.plot(plants_simple["Time"], plants_simple["Lateral_spread"], label="Lateral spread")
plt.title("Plant height and lateral spread during growth")
plt.legend()
```

```python
# Vi kan namnge axlarna med xlabel och ylabel.
plt.plot(plants_simple["Time"], plants_simple["Plant_height"], color="g", linewidth=2, linestyle="dotted", label="Plant height")
plt.plot(plants_simple["Time"], plants_simple["Lateral_spread"], label="Lateral spread")
plt.title("Plant height and lateral spread during growth")
plt.xlabel("Time")
plt.ylabel("Centimers")
plt.legend()
```

```python
# Vi ange färg på våra "tick-labels" och vrida dem med tick_params(). Du kan testa direction, labelsize och labelrotation.
plt.plot(plants_simple["Time"], plants_simple["Plant_height"], color="g", linewidth=2, linestyle="dotted", label="Plant height")
plt.plot(plants_simple["Time"], plants_simple["Lateral_spread"], label="Lateral spread")
plt.title("Plant height and lateral spread during growth")
plt.xlabel("Time (days)")
plt.ylabel("Centimers")
plt.tick_params(axis="x", direction="out", labelsize="large", labelrotation=30, )
plt.grid()
plt.legend()
```

```python
# Till sist sparar vi via plt.savefig(). bbox_incehs bestämmer hur mycket marginal det ska finnas.
plt.plot(plants_simple["Time"], plants_simple["Plant_height"], color="g", linewidth=2, linestyle="dotted", label="Plant height")
plt.plot(plants_simple["Time"], plants_simple["Lateral_spread"], label="Lateral spread")
plt.title("Plant height and lateral spread during growth")
plt.xlabel("Time (days)")
plt.ylabel("Centimers")
plt.tick_params(axis="x", direction="out", labelsize="large", labelrotation=30)
plt.legend()
plt.savefig("my_fig.png", bbox_inches=None)
```
