```python
# Vi importerar datasettet och plt.
import pandas as pd
import matplotlib.pyplot as plt

fires = pd.read_csv("fires.csv")
fires = fires.head(10)
```

```python
# Bar-charts tar in x, height, width, align (och andra som color)
plt.figure(figsize=(10, 6))
plt.bar(x=fires["state"], height=fires["firespots"], align="edge", width=0.8, color="r")
```

```python
# Scatterplot anv√§nds ofta f√∂r att j√§mf√∂ra korrelationer.
# Du kan anv√§nda alpha och s f√∂r att g√∂ra det tydligare.
spotify_stats = pd.read_csv("spotify_data_by_genres.csv")

plt.scatter(spotify_stats["danceability"], spotify_stats["valence"], alpha=0.5, s=20)
```

```python
# SUBPLOTS
# Initieras med plt.subplots(). Konventionen √§r fig och axs.
# Du styr varje subplot med axs[i, j]. M√•nga egenskaper byter namn till set_x
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].scatter(spotify_stats["popularity"], spotify_stats["valence"], alpha=0.5, s=20)
axs[0, 0].set_title("Valence")
axs[0, 0].set_xlabel("Popularity")
axs[0, 0].set_ylabel("Valence")

axs[0, 1].scatter(spotify_stats["popularity"], spotify_stats["loudness"], alpha=0.5, s=20)
axs[0, 1].set_title("Loudness")
axs[0, 1].set_xlabel("Popularity")
axs[0, 1].set_ylabel("Loudness")

axs[1, 0].scatter(spotify_stats["popularity"], spotify_stats["duration_ms"], alpha=0.5, s=20)
axs[1, 0].set_title("Duration (ms)")
axs[1, 0].set_xlabel("Popularity")
axs[1, 0].set_ylabel("Duration (ms)")

axs[1, 1].scatter(spotify_stats["popularity"], spotify_stats["popularity"], alpha=0.5, s=20)
axs[1, 1].set_title("Popularity")
axs[1, 1].set_xlabel("Popularity")
axs[1, 1].set_ylabel("Valence")
fig.suptitle("Popularity and ___")
```

```python
# Cirkeldiagram (Pie chart)
# Tar in x (summan som ska r√§knas om till en andel) och en m√§rkning p√• varje skiva.
market_share = pd.read_csv("EV_market_share.csv")

plt.pie(x=market_share["number_sold"], labels=market_share["model"])
```

```python
# Histogram visar ofta f√∂rdelningen f√∂r _en_ viss variabel

countries = pd.read_csv("countries.csv")

plt.hist(x=countries["gdp"], bins=20)
```
